```python
import wfdb

# 1. Listar registros
records = wfdb.get_record_list('emgdb')
print("Registros disponibles:", records)

# 2. Elegir uno
record_name = records[1]

# 3. Descargar y leer señal
record = wfdb.rdrecord(record_name, pn_dir='emgdb/1.0.0')

signal = record.p_signal
fs = record.fs

print("Shape:", signal.shape)
print("Fs:", fs)
```

```python
import matplotlib.pyplot as plt

plt.plot(signal)
plt.title("Señal EMG")
plt.grid(ls=":")
plt.xlabel("Muestras")
plt.ylabel("Amplitud")
plt.xlim(6000,8000)
plt.show()
```
