```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
A1, A2, A3 = 5, 6, 7
phi1, phi2, phi3 = 0, np.pi/2, np.pi*2
f1, f2, f3 = 200, 150, 100
w1 = 2*np.pi*f1
w2 = 2*np.pi*f2
w3 = 2*np.pi*f3

fs = 20 * max(f1, f2, f3)
Ts = 1/fs

n = np.arange(2000)
t = n * Ts

x1 = A1*np.sin(w1*t+phi1)
x2 = A2*np.sin(w2*t+phi2)
x3 = A3*np.sin(w3*t+phi3)
x = x1+x2+x3
```

```python
plt.figure(figsize=(8,6))
plt.plot(t,x)
plt.xlim(0,0.1)
```

```python
plt.stem(n,x)
plt.xlim(0,100)
```

```python
plt.magnitude_spectrum(x, Fs=fs, scale='dB')
plt.grid(ls=":")
plt.xlim(0,400)
plt.show()
```

```python
f, tt, Zxx = signal.stft(x, fs=fs, nperseg=256, noverlap=128)

# Graficar
plt.figure(figsize=(15, 4))
plt.pcolormesh(tt, f, np.abs(Zxx), shading='gouraud')

plt.title('Espectrograma de la STFT')
plt.ylabel('Frecuencia [Hz]')
plt.xlabel('Tiempo [s]')
plt.ylim(0, 350)   # subimos un poco para ver bien 300 Hz
plt.colorbar(label='Magnitud')
plt.tight_layout()
plt.show()
```
